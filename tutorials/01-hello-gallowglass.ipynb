# Hello Gallowglass

Gallowglass is a typed functional language compiling to PLAN. This notebook walks through the core: declarations, types, algebraic types, and pattern matching. Cells run in this kernel sequentially, so later cells see earlier definitions.

You'll need the Gallowglass kernel installed:

```bash
python3 -m bootstrap.jupyter_kernel install
```

Each code cell below is one input to the kernel. Declaration cells echo a one-line summary (`name : Type`); expression cells echo the result.

## Functions and types

Define a function with `let`. Every binding has a type — the compiler infers types when you don't spell them out.

````gallowglass
let twice : Nat → Nat = λ n → n + n
````

Apply the function. The cell's value is the displayed result; the renderer picks colours based on the type (numbers in cyan, constructors in bold blue, strings in green).

````gallowglass
twice 21
````

When you omit the type annotation, the compiler infers one and the cell summary shows what it picked:

````gallowglass
let triple = λ n → n * 3
````

````gallowglass
triple 7
````

## Algebraic types

Define your own data shapes with `type`. Each constructor is a value (or a constructor function, if it takes arguments).

````gallowglass
type Coin =
  | Heads
  | Tails
````

````gallowglass
Heads
````

Bring a type from the prelude into scope with `use … unqualified { … }`. `Pair a b` is the standard product type — its single constructor is `MkPair`.

````gallowglass
use Core.Pair unqualified { Pair, MkPair }
````

````gallowglass
let pair_val : Pair Nat Nat = MkPair 3 7
````

````gallowglass
pair_val
````

## Pattern matching

`match` dispatches on a constructor. Each arm names the constructor and binds its fields:

````gallowglass
let coin_value : Coin → Nat
  = λ c → match c {
      | Heads → 1
      | Tails → 0
    }
````

````gallowglass
coin_value Heads
````

````gallowglass
coin_value Tails
````

Pattern arms can bind fields. Note: single-letter lowercase identifiers are reserved as type variables, so we use `aa` and `bb` here rather than `a` and `b`:

````gallowglass
let sum_pair : Pair Nat Nat → Nat
  = λ pp → match pp { | MkPair aa bb → aa + bb }
````

````gallowglass
sum_pair pair_val
````
